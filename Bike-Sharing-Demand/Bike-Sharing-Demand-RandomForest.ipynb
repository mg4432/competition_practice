# load data

```python
import pandas as pd 
import numpy as np 

train = pd.read_csv('dataset/train.csv')
test = pd.read_csv('dataset/test.csv')
submission = pd.read_csv('dataset/sampleSubmission.csv')
```

# Feature Engineering 

weather : 4인 데이터 삭제

```python
train = train[train['weather'] != 4]
```

## concat train/test
- 같은 피처 엔지니어링을 적용하기 위해 train/test set을 합쳐줌

```python
all_data_temp = pd.concat([train, test])
all_data_temp
```

인덱스 번호가 제대로 부여되지 않음 
> `ignore_index=True` 옵션을 이용

```python
all_data = pd.concat([train, test], ignore_index = True)
all_data
```

```python
del all_data_temp
```

## 파생변수 추가
- year, month, hour, weekday 변수만 사용

```python
from datetime import datetime
```

```python
# date
all_data['date'] = all_data['datetime'].apply(lambda x : x.split()[0])

# year
all_data['year'] = all_data['datetime'].apply(lambda x : x.split()[0].split('-')[0])

# month
all_data['month'] = all_data['datetime'].apply(lambda x : x.split()[0].split('-')[1])

# hour
all_data['hour'] = all_data['datetime'].apply(lambda x : x.split()[1].split(':')[0])

# weekday
all_data['weekday'] = all_data['date'].apply(
    lambda datestring : datetime.strptime(datestring, '%Y-%m-%d').weekday())
```

## 필요없는 피처 제거
- casual, registered, datetime, date, windspeed, month

```python
all_data
```

```python
drop_features = ['casual', 'registered', 'datetime', 'date', 'windspeed', 'month']
all_data.drop(drop_features, axis = 1, inplace = True)
```

## train/test split

```python
all_data
```

```python
X_train = all_data[~all_data['count'].isnull()]
X_test = all_data[all_data['count'].isnull()]
```

```python
X_test
```

```python
X_train = X_train.drop(['count'], axis = 1) 
X_test = X_test.drop(['count'], axis = 1)

y = train['count']
```

## 평가지표 함수 작성

```python
def rmsle(y_true, y_pred, convertExp = False) : 
    # 로그변환을 했다는 가정 하에 지수 변환으로 원래 값으로 변환
    if convertExp :
        y_true = np.exp(y_true)
        y_pred = np.exp(y_pred)
    

    log_true = np.log1p(y_true)
    log_pred = np.log1p(y_pred)
    # RSMLE 계산
    output = np.sqrt( np.mean( (log_true - log_pred) ** 2 ) )
    return output
```

## 훈련

**실제로는 훈련 데이터로 성능을 검증하면 안되지만, 단순히 구현해보기 위한 과정임**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn import metrics 

rf_model = RandomForestRegressor()
```

```python
rmsle_scorer = metrics.make_scorer(rmsle, greater_is_better = False)


rf_params = {'random_state' : [42],
             'n_estimators' : [50, 80, 100,120,140],}

gridsearch_rf_model = GridSearchCV(estimator = rf_model,
                                   param_grid = rf_params,
                                   scoring = rmsle_scorer,
                                   cv = 5)
```

```python
log_y = np.log(y)
gridsearch_rf_model.fit(X_train, log_y)
```

```python
print('최적 하이퍼 파라미터 : ', gridsearch_rf_model.best_params_)
```

```python
# 예측 
preds = gridsearch_rf_model.best_estimator_.predict(X_train)

# 평가
print(f'랜덤 포레스트 회귀 RMSLE 값 : {rmsle(log_y, preds, True):.4f}')
```

## 예측 및 결과 제출 

- 1. 테스트 데이터로 예측한 결과를 제출해야 함 
- 2. 예측한 값에 지수변환을 해줘야 함 

```python
rf_reg_preds = gridsearch_rf_model.best_estimator_.predict(X_test)

submission['count'] = np.exp(rf_reg_preds)
submission.to_csv('submission.csv', index = False)
```
